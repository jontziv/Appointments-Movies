Grouping and Aggregating DataFrames

Aggregate the appointments Dataset and create a patients dataset (variable name: patients) with one row for each unique patient. The new patients dataset should include the following information for each patient:<br>
- Age
- Scholarship
- Hipertension
- Diabetes
- Alcoholism
- Handcap
- mean TimeGapDays
- Gender
- number of Appointments 
- number of SMS_received
- number of No-shows
- No-show rate (calculate separately based on number of No-shows)

```python
# Group by 'PatientId' and aggregate the information for each patient
patients = df.groupby('PatientId').agg({
    'Age': 'mean',
    'Scholarship': 'first',
    'Hipertension': 'first',
    'Diabetes': 'first',
    'Alcoholism': 'first',
    'Handcap': 'first',
    'TimeGapDays': 'mean',
    'Gender_M': 'first',
    'AppointmentID': 'count',
    'SMS_received': 'sum',
    'No-show': 'sum',
}).reset_index()
```

```python
patients
```

```python
# Rename columns for clarity
patients.rename(columns={
    'Age': 'Mean_Age',
    'Scholarship': 'Scholarship',
    'Hipertension': 'Hypertension',
    'Diabetes': 'Diabetes',
    'Alcoholism': 'Alcoholism',
    'Handcap': 'Handicap',
    'TimeGapDays': 'Mean_TimeGapDays',
    'Gender_M': 'Gender',
    'AppointmentID': 'Number_of_Appointments',
    'SMS_received': 'Number_of_SMS_received',
    'No-show': 'Number_of_No_shows',
}, inplace=True)
```

```python
patients
```

```python
# Calculate the No-show rate based on the number of appointments and no-shows
patients['No_show_rate'] = patients['Number_of_No_shows'] / patients['Number_of_Appointments']
```

Some observations have the non-intuitive value -1 in the column TimeGapDays which is the time gap between ScheduledDay and AppointmentDay. Identify the reason for getting the value -1 and propose an alternative way to calculate the timegap without negative values. 

```python
df
```

```python
# Convert 'ScheduledDay' and 'AppointmentDay' columns to datetime objects explicitly
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'], utc=True)
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'], utc=True)

# Calculate the time gap without negative values
df['TimeGapDays'] = (df['AppointmentDay'].dt.floor('D') - df['ScheduledDay'].dt.floor('D')).dt.days
```

```python
# Define a function to remove special characters from column names
def remove_special_characters(column_name):
    # Use regex to remove all characters that are not letters or digits
    return ''.join(e for e in column_name if e.isalnum())

# Rename columns by applying the function to each column name
df.rename(columns=remove_special_characters, inplace=True)

# Display the DataFrame with updated column headers
print(df)
```
